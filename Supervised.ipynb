Impor:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

Membaca Dataset Yang Sudah Dibersihkan:
- 
- Kita membaca dataset yang telah dibersihkan dari file CSV menggunakan pd.read_csv().
- Kemudian, kita menampilkan 5 baris pertama dari dataset untuk mendapatkan gambaran umum tentang struktur dan isi data.

```python
# Membaca dataset yang sudah dibersihkan
cleaned_file_path = "cleaned_student_mat.csv"  # Ganti dengan path file Anda
cleaned_df = pd.read_csv("cleaned_student_mat.csv")

# Menampilkan 5 baris pertama dataset
print(cleaned_df.head())
```

Menampilkan Informasi Dataset:
-
- cleaned_df.info(): Menampilkan informasi tentang jumlah entri, kolom, dan tipe data dari setiap kolom.
- cleaned_df.describe(): Menghasilkan statistik deskriptif untuk kolom numerik, seperti rata-rata, standar deviasi, nilai minimum dan maksimum, serta persentil.

```python
# Informasi dataset
cleaned_df.info()

# Statistik deskriptif
print(cleaned_df.describe())
```

Menampilkan Data dengan sample 10 data:
-
- Menggunakan cleaned_df.sample(10) untuk menampilkan 10 baris acak dari dataset. Ini memberikan pandangan yang lebih luas tentang data yang ada.

```python
# Menampilkan 10 baris acak
print(cleaned_df.sample(10))
```

Memvisualisasi Dataset:
-
- Membuat histogram untuk setiap kolom numerik dalam dataset. Histogram ini membantu kita memahami distribusi nilai dari setiap fitur.

```python
# Visualisasi histogram untuk kolom numerik
cleaned_df.hist(figsize=(15, 10), bins=20)
plt.suptitle('Distribusi Data Setelah Dibersihkan')
plt.show()
```

Visualisasi Dsitribusi NIlai Akhir (G3):
- 
- Menggunakan sns.histplot() untuk membuat histogram dari kolom nilai akhir (G3) dengan kurva kepadatan (KDE). Ini memberikan gambaran yang lebih jelas tentang distribusi nilai akhir siswa.

```python
# Visualisasi distribusi G3
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_df['G3'], bins=20, kde=True)
plt.title('Distribusi Nilai Akhir (G3)')
plt.xlabel('G3')
plt.ylabel('Frekuensi')
plt.show()
```

Analisis Korelasi:
-
- Mengonversi kolom G3 ke tipe numerik jika belum.
- Memilih hanya kolom dengan tipe data numerik untuk analisis korelasi.
- Menggunakan sns.heatmap() untuk memvisualisasikan matriks korelasi, yang menunjukkan hubungan antara fitur-fitur numerik dalam dataset.

```python
# Mengonversi kolom yang diperlukan ke tipe numerik
cleaned_df['G3'] = pd.to_numeric(cleaned_df['G3'], errors='coerce')

# Memilih hanya kolom numerik untuk matriks korelasi
numeric_df = cleaned_df.select_dtypes(include=['number'])

# Matriks Korelasi
plt.figure(figsize=(12, 10))
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```

Persiapan Data Untuk Klasifikasi:
-
- Memisahkan fitur (X) dari variabel target (y). Di sini, kita menetapkan y sebagai 1 jika G3 lebih besar dari 10 (lulus), dan 0 jika tidak.
- Mengonversi variabel kategorikal menjadi variabel dummy menggunakan pd.get_dummies().
- Membagi dataset menjadi data latih dan data uji dengan proporsi 70:30 menggunakan train_test_split().

```python
# Menyiapkan data untuk klasifikasi
X = cleaned_df.drop(['G3'], axis=1)  # Fitur
y = (cleaned_df['G3'] > 10).astype(int)  # Variabel target (1 jika lulus, 0 jika gagal)

# Mengonversi variabel kategorikal menjadi variabel dummy
X = pd.get_dummies(X, drop_first=True)

# Membagi data menjadi data latih dan data uji
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Melatih Model Klasifikasi:
-
- Mengimpor RandomForestClassifier dari sklearn.ensemble.
- Membuat dan melatih model Random Forest menggunakan data latih.
- Menggunakan model untuk memprediksi kelas pada data uji.

```python
from sklearn.ensemble import RandomForestClassifier

# Melatih model Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Prediksi
y_pred = rf_model.predict(X_test)
```

Evaluasi Model Klasifikasi:
-
- Menggunakan classification_report untuk menampilkan metrik evaluasi seperti precision, recall, dan f1-score untuk setiap kelas.
- Menghitung dan mencetak akurasi keseluruhan model menggunakan accuracy_score().

```python
from sklearn.metrics import classification_report, accuracy_score

# Evaluasi untuk klasifikasi
print(classification_report(y_test, y_pred))
print("Akurasi:", accuracy_score(y_test, y_pred))
```

Melaukan Persiapan Data Untuk Regresi:
-
- Memisahkan fitur (X_reg) dari variabel target (y_reg) untuk model regresi. Di sini, y_reg adalah nilai akhir (G3).
- Mengonversi variabel kategorikal menjadi variabel dummy.
- Membagi dataset menjadi data latih dan data uji untuk regresi.

```python
# Menyiapkan data untuk regresi
X_reg = cleaned_df.drop(['G3'], axis=1)  # Fitur
y_reg = cleaned_df['G3']  # Variabel target (G3)

# Mengonversi variabel kategorikal menjadi variabel dummy
X_reg = pd.get_dummies(X_reg, drop_first=True)

# Membagi data menjadi data latih dan data uji
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)
```

Melakukan Model Regresi:
-
- Mengimpor LinearRegression dari sklearn.linear_model.
- Membuat dan melatih model regresi linier menggunakan data latih.
- Menggunakan model untuk memprediksi nilai G3 pada data uji.

```python
from sklearn.linear_model import LinearRegression

# Melatih model regresi linier
lr_model = LinearRegression()
lr_model.fit(X_train_reg, y_train_reg)

# Prediksi
y_pred_reg = lr_model.predict(X_test_reg)
```

Evaluasi Model Regresi:
-
- Menggunakan mean_squared_error untuk menghitung kesalahan kuadrat rata-rata antara nilai yang diprediksi dan nilai aktual.
- Menghitung R Score untuk mengevaluasi seberapa baik model menjelaskan variabilitas data.

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluasi untuk regresi
print("Mean Squared Error:", mean_squared_error(y_test_reg, y_pred_reg))
print("R Score:", r2_score(y_test_reg, y_pred_reg))
```
